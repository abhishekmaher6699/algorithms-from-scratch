```python
import numpy as np
```

```python
class LogisticRegression():
    def __init__(self, learning_rate=0.1, n_iters=1000):
      self.lr = learning_rate
      self.n_iters = n_iters
      self.beta0 = None
      self.betas = None

    def sigmoid(self, z):
      return 1/(1 + np.exp(-z))

    def loss_function(self, pred, y):
      loss = (-1/ len(y)) * np.sum(y*np.log(pred) - (1-y)*np.log(1-pred))
      return loss

    def fit(self, X, y):
      X = np.c_[np.ones((X.shape[0], 1)), X]

      m = X.shape[0] # rows
      n = X.shape[1] #features

      self.betas = np.zeros(n)
      self.beta0 = 0

      for i in range(self.n_iters):
        z = np.dot(X, self.betas) + self.beta0
        pred = self.sigmoid(z)
        loss = self.loss_function(pred, y)
        # print(loss)
        error = pred - y

        dw = (1 / m) * np.dot(X.T, error)
        db = (1 / m) * np.sum(error)

        self.beta0 = self.beta0 - self.lr*db
        self.betas = self.betas - self.lr*dw

    def predict(self,X):
        X = np.c_[np.ones((X.shape[0], 1)), X]
        z = self.beta0 + np.dot(X, self.betas)
        pred = self.sigmoid(z)

        return pred
```

```python
X_train = np.array([[2,5], [3,5], [5,0], [6,7], [8,0]])
y_train = np.array([0, 0, 0, 1, 1])
```

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)
lr.predict(X_train)
```
